## Style sheet

`style` package add styles. You can use pre-defined styles.

````python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from cycler import cycler
plt.style.use('ggplot')
data = np.random.randn(50)
````

Now we can plot using the `ggplot` style

````python
fig, ax = plt.subplots()
ax.plot(data)
````

## Define a custom style

You can create a style and use it by calling `style.use` with the path to the style sheet.

### Save your custom style

If you add your `<style-name>.mplstyle` file to `mpl_configdir/stylelib`, you can reuse your custom style sheet with a call to `style.use(<style-name>)`. By default `mpl_configdir` should be `~/.config/matplotlib`, but you can check where yours is with `matplotlib.get_configdir()`; you may need to create this directory. 

For example you can create a presentation style in the file `mpl_configdir/stylelib/presentation.mplstyle`:

```python
axes.titlesize : 24
axes.labelsize : 20
lines.linewidth : 3
lines.markersize : 10
xtick.labelsize : 16
ytick.labelsize : 16
```

and then use it to make automatically your plot adapt for a presentation by adding:

```python
>>> import matplotlib.pyplot as plt
>>> plt.style.use('presentation')
```

## Compose styles

Style sheets are designed to be composed together. So you can have a style sheet that customizes colors and a separate style sheet that alters element sizes for presentations. These styles can easily be combined by passing a list of styles:

```python
>>> import matplotlib.pyplot as plt
>>> plt.style.use(['dark_background', 'presentation'])
```

Note that styles further to the right will overwrite values that are already defined by styles on the left.

## Temporary style

If you want to use a specific style only for a specific plot:

````python
with plt.style.context('dark_background'):
    plt.plot(np.sin(np.linspace(0, 2 * np.pi)), 'r-o')
plt.show()
````

## Changing rc settings

It's possible to dynamically change the default rc settings. They're stored in a dictionary-like variable called `matplotlib.rcParams`

````python
mpl.rcParams['lines.linewidth'] = 2
mpl.rcParams['lines.linestyle'] = '--'
plt.plot(data)
````

`matplotlib.rc()` can be used to modify multiple settings at once, using keyword

````python
mpl.rc('lines', linewidth=4, linestyle='-.')
plt.plot(data)
````

Then the default settings can be restored with `matplotlib.rcdefaults()`
